# 電極電位

## 標準電極電位 $E^\circ$, Standard electrode potential
$\require{mhchem}$

\begin{align*}
\ce{O} + n \ce{e- <=> R}&
\end{align*}

- ${\rm O}$: oxidant/酸化体(電子を受け取る)
- ${\rm R}$: reductant/還元体(電子を受け取った後)

この反応の電位で, ${\rm O/R}$が 基準状態($a=1$)となる電位を  **標準電極電位** $E^\circ_{\ce{(O/R)}}$

## 電位の原点
(負の電荷を持つ)電子$\ce{e-}$が上の反応で $E^\circ_{\rm O/R}$ の影響を受けるので, 電子授受平衡を考えると
\begin{align*}
\Delta_{\rm f}G^\circ(\ce{O}) + (-n\mathcal{F})E^\circ_{\ce{(O/R)}} 
= \Delta_{\rm f}G^\circ(\ce{R})
\end{align*}
この反応を利用して, 電位測定の基準となる電位を周期律ようで最も軽い元素を利用した
水素分子 $\ce{H2}$ と 水素イオン $\ce{H+}$ を利用して定義する.  
水素の標準生成ギブスエネルギー$\Delta_{\rm f} G(\ce{H2})$ は中性単体($\ce{h2}$)では(単体で安定しているので) ゼロ と考えました. さらに水素イオン $\ce{H+}$ の $\Delta_{\rm f} G(\ce{H+})$ もゼロにとると. $\ce{O=2H+, R=H2}$ のとき $E^\circ_{\ce{2H+/H2}} =E^\circ=0$

標準**水素**電極電位(Standard Hydrogen Electrode) では次の反応の電位を基準(ゼロ)として他の電位を相対的に決めます.
\begin{align*}
\ce{2H+} + \ce{2e- <=> H2}&
\end{align*}
この定義は $\ce{H+}$ で $a_{\ce{H+}}=1$($[\ce{H+}]={\rm 1 mol/L}, {\rm pH=0}$) の水溶液に白金電極を利用した装置で測定できます.

$\ce{H2}$の燃焼反応では
\begin{align*}
&\ce{2H+ + 2e- <=> H2}&\\
&\ce{1/2O2 + 2H+ + 2e- <=> H2O}&
\end{align*}

上段の式が $E^\circ=0,$ 下段の式が$E^\circ=+1.23{\rm \ V}$ となり, $E^\circ = {\rm +1.23 V} vs. \text{SHE}$ と表現できます.
