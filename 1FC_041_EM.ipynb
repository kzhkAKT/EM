### 比電気抵抗/比抵抗率 $\rho\ ({\rm \Omega m})$

抵抗 $R\ ({\rm \Omega})$

オームの法則 
\begin{align*}
R &= \dfrac{\Delta E}{I}
\end{align*}

比抵抗率 $\rho\ ({\rm \Omega m})$
\begin{align*}
R &\propto l&\\
R &\propto \dfrac{1}{A}&\\
R &= \rho \dfrac{l}{A} &
\end{align*}

#### 電位電導率/電動率 $\kappa\ ({\rm S m^{-1}})$

比抵抗率 $\rho$ の逆数
\begin{align*}
\kappa &=  \dfrac{1}{\rho}
\end{align*}
$\kappa$ はギリシャ文字で kappa とよみます.
